```python
#Solution in Python
nums = list(map(int, input().split()))
target = int(input())
res = []
for i in range(len(nums)-1):
    for j in range(i+1,len(nums)):
        if(nums[i]+nums[j]==target):
            res.append(i)
            res.append(j)
res[:2]
```
